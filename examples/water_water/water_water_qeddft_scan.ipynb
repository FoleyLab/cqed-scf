# Water dimer CQED-DFT scan

Run this notebook from the `p4dev` conda environment, which should provide Psi4 and the editable `cqed-scf` package. The first two imports are standard for a Psi4/NumPy workflow. The package-specific import is `CQEDCalculator`; it is the high-level cqed-scf wrapper used below to run cavity-modified SCF/DFT calculations while still relying on Psi4-style molecular input and options.

```python
import psi4
import numpy as np
from cqed_scf.calculator import CQEDCalculator
```

## Calculator setup

`PSI4_OPTIONS` is intentionally familiar Psi4 input: basis, RKS reference, density fitting, convergence thresholds, and DFT grid. The cqed-scf-specific inputs start with `LAMBDA_VECTOR` and `OMEGA`. The three components of `LAMBDA_VECTOR` set the cavity-coupling direction and strength in the molecular Cartesian frame, while `OMEGA` is the photon frequency in atomic units. Passing these values into `CQEDCalculator` creates a reusable calculator; later calls to `calc.energy(...)` are the cqed-scf analogue of calling `psi4.energy(...)` for each geometry. The `functional` argument controls the DFT functional used inside the cqed-scf driver, with dispersion handled by the package when a supported dispersion-corrected functional such as `wb97x-d` is requested.

```python
# -----------------------------------------------------------------------------
# QED-DFT settings
# -----------------------------------------------------------------------------

PSI4_OPTIONS = {
    "basis": "6-311G*",
    "reference": "rks",
    "scf_type": "df",
    "e_convergence": 1e-9,
    "d_convergence": 1e-9,
    "dft_radial_points": 99,
    "dft_spherical_points": 590,
    "dft_pruning_scheme": "none",
}

LAMBDA_VECTOR = np.array([0.078, 0.055, 0.027])
OMEGA = 0.06615
FUNCTIONAL = "wb97x-d"

#----------------------------------------------------
# Create cqed-scf calculator with the settings above
#----------------------------------------------------
calc = CQEDCalculator(
    lambda_vector = LAMBDA_VECTOR,
    psi4_options = PSI4_OPTIONS,
    omega = OMEGA,
    charge = 0,
    multiplicity = 1,
    density_fitting = True,
    functional = FUNCTIONAL,
    debug = False
)

# Helpful constant
HARTREE_TO_KCAL_MOL = 627.509474

```

## Running the CQED-DFT scan

This loop uses Psi4 for the geometry bookkeeping: each dimer string becomes a Psi4 molecule, and `extract_subsets` builds the two monomer molecules from the same dimer geometry. The cqed-scf-specific step is the call to `calc.energy(...)`. `CQEDCalculator.energy` expects a Psi4-format geometry string, then internally runs the CQED-SCF/CQED-DFT calculation using the Psi4 options, `lambda_vector`, photon frequency, functional, charge, and multiplicity configured above. Each row of `qed_dft_energies` stores the resulting dimer, monomer A, and monomer B CQED-DFT total energies.

```python
# In the following loop, we are going to go through GEOMETRIES list
# pass each geometry string to psi4.geometry to make a psi4 molecule representing the dimer
# use extract_subset to make psi4 molecules representing each monomer
# and then use psi4.create_psi4_string_from_molecule() to get strings for each monomer
# We will then calculate the QED-DFT energy using our calc instance above, which will expect a 
# string representing the geometry


qed_dft_energies = np.zeros((8,3)) #<-- will hold dimer, monomer_a, and monomer_b qed-dft energies
index = 0
for geom_str in GEOMETRIES:
    dimer = psi4.geometry(geom_str)
    monomer_a = dimer.extract_subsets(1,2)
    monomer_b = dimer.extract_subsets(2,1)
    dimer_str = dimer.create_psi4_string_from_molecule()
    monomer_a_str = monomer_a.create_psi4_string_from_molecule()
    monomer_b_str = monomer_b.create_psi4_string_from_molecule()
    qed_dft_energies[index, 0] = calc.energy(dimer_str)
    qed_dft_energies[index, 1] = calc.energy(monomer_a_str)
    qed_dft_energies[index, 2] = calc.energy(monomer_b_str)
    index += 1
    
```

## Interaction energies

The interaction energy is assembled in the usual supermolecular way, `E_dimer - E_monomer_a - E_monomer_b`. The important cqed-scf point is that all three energies came from `CQEDCalculator.energy`, not plain `psi4.energy`, so the result is the water-dimer interaction curve for the selected cavity coupling and photon frequency. The printed values are in hartree; multiply by `HARTREE_TO_KCAL_MOL` if you want kcal/mol.

```python
# define interaction energies
E_int = qed_dft_energies[:,0] - qed_dft_energies[:,1] - qed_dft_energies[:,2]

# convert to kcal / mol
E_int *= HARTREE_TO_KCAL_MOL

print(E_int)
```
